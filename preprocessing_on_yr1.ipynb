```python
import pandas as pd
from sklearn import preprocessing
data = pd.read_csv('survey_fcasts.yr1.tab', sep='\t', lineterminator='\n')
events = pd.read_csv('ifps.csv', encoding = "ISO-8859-1")
```

```python
closed = events['q_status']=='closed'
binary = events['n_opts'] == 2
events = events[closed & binary]
```

```python
merged = pd.merge(data, events, left_on=['ifp_id','answer_option'], right_on=['ifp_id','outcome'], how='inner')
```

```python
merged['mse'] = (1-merged['value'])**2
```

```python
ifp = merged.groupby('ifp_id', as_index=False)['mse'].mean()
ifp.columns = ['ifp_id', 'ifp_mean']
ifp['ifp_std'] = merged.groupby('ifp_id')['mse'].std().reset_index()['mse']
ifp.head()
```

```python
merged = pd.merge(merged, ifp, left_on=['ifp_id'], right_on=['ifp_id'], how='inner')
```

```python
merged['smse']= (merged['mse']-merged['ifp_mean'])/merged['ifp_std']
merged[['ifp_id', 'ctt','cond','training','user_id','value','smse']].head()
```

```python
user = merged.groupby(['user_id','cond','training'],as_index=False)['smse'].mean()
merged.groupby(['ctt'],as_index=False)['smse'].mean()
```
