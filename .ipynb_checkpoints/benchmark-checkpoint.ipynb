```python
import numpy as np
import itertools as it

def data_generation(n_cat, n_cat_efct):
    n_cell = 2**n_cat
    cell_size = 2 * np.ones(n_cell, dtype=int)
    n_sample = np.sum(cell_size)
    mu = np.array([i for i in range(2**n_cat_efct)])
    print(mu)
    
    x = np.zeros((n_sample, n_cat))
    y = np.zeros((n_sample, ))
    cat = np.zeros(n_sample)
    idx = 0
    for i, v in enumerate(it.product([0,1], repeat = n_cat)):
        for j in range(cell_size[i]):
            cat[idx] = i
            x[idx, :] = v
            y[idx] = mu[i >> (n_cat - n_cat_efct)]
            idx += 1

    idx = np.arange(n_sample)
    np.random.shuffle(idx)
    cat = cat[idx]
    x = x[idx]
    y = y[idx]
    return cat, x, y

cat, x, y = data_generation(5, 3)
for i in range(len(x)):
    
    
    print(x[i, :], y[i])
    print(cat[i])

```
